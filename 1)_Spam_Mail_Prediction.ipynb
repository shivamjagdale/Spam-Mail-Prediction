```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv("/content/drive/MyDrive/ML Datasets/Mail-Data.csv")
df = pd.DataFrame(a)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(" ")
```

```python
df["Category"].value_counts()
```

```python
df["Category"] = df["Category"].replace("spam", 0)
df["Category"] = df["Category"].replace("ham", 1)
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df["Message"]
```

```python
Y = df["Category"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify=Y, shuffle=True)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
vectorizer = TfidfVectorizer()
```

```python
vectorizer.fit(X_train)
```

```python
X_train = vectorizer.transform(X_train)
```

```python
X_test = vectorizer.transform(X_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_prediction = model.predict(X_train)
X_train_acc = accuracy_score(X_train_prediction, Y_train)
print(X_train_acc)
```

```python
X_test_prediction = model.predict(X_test)
X_test_acc = accuracy_score(X_test_prediction, Y_test)
print(X_test_acc)
```
